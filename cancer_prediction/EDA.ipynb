## Model Deployment- Ensembling Techniques


```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
rfc=RandomForestClassifier()
abc=AdaBoostClassifier()
classifiers=[rfc,abc]
```

##### As data is imbalanced use stratified kfold - that represents equal proportion of data groups
##### Evaluating the model's performance using precision_score,recall_score,f1_score

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
x=copy_df.drop(['diagnosis'],axis=1)
y=copy_df['diagnosis']

skf=StratifiedKFold(n_splits=5)

rfc_score={'precision':[],'recall':[],'f1_score':[]}
abc_score={'precision':[],'recall':[],'f1_score':[]}

skf.get_n_splits(x,y)
evaluators=[precision_score,recall_score,f1_score]

for train_index,test_index in skf.split(x,y):
    x_train,x_test=x.iloc[train_index],x.iloc[test_index]
    y_train,y_test=y.iloc[train_index],y.iloc[test_index]
    for classifier in classifiers:
        classifier.fit(x_train,y_train)
        prediction=classifier.predict(x_test)
        for evaluator in evaluators:
            if classifier==rfc:
                if evaluator==precision_score:
                    rfc_score['precision'].append(precision_score(prediction,y_test))
                elif evaluator==recall_score:
                    rfc_score['recall'].append(recall_score(prediction,y_test))
                elif evaluator==f1_score:
                    rfc_score['f1_score'].append(f1_score(prediction,y_test))
                   
            else:
                if evaluator==precision_score:
                    abc_score['precision'].append(precision_score(prediction,y_test))
                elif evaluator==recall_score:
                    abc_score['recall'].append(recall_score(prediction,y_test))
                elif evaluator==f1_score:
                    abc_score['f1_score'].append(f1_score(prediction,y_test))
print(rfc_score['precision'])
print(abc_score['precision'])
```
